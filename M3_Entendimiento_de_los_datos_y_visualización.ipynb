## <strong>Visualizar las estadísticas</strong>

<p>Este conjunto de datos es pequeño, pero si se quisiera saber la cantidad de atributos y de elementos que se tienen en el conjunto de datos, se puede utilizar la función <code>dataframe.shape</code>. Esta función visualiza primero el número de elementos y luego el número de atributos.</p>

```python
df.shape
```

<p>Se va a utilizar la función <code>dataframe.describe</code> para visualizar las estadísticas del conjunto de datos. Por defecto, la función <code>dataframe.describe</code> muestra las columnas numéricas y mostrará los siguientes datos:</p>
<ul>
    <li>el recuento de esa variable</li>
    <li>la media</li>
    <li>la desviación estándar (std)</li> 
    <li>el valor mínimo</li>
    <li>los cuartiles </li>
    <li>el valor máximo</li>

```python
df.describe()
```

Por defecto la función solo muestra atributos que son numéricos. Es posible hacer que la función <code>describe</code> funcione también para las columnas de tipo object. Para permitir un resumen de todas las columnas, podríamos añadir un argumento <code>include="object"</code> o <code>include="all"</code> entre los paréntesis de la función <code>describe</code>.

```python
df.describe(include=['object'])
```

## <strong>Explorar relaciones entre los datos</strong>

<b>Variables numéricas:</b>

<p>Las variables numéricas pueden ser continuas o discretas. Las variables numéricas pueden tener el tipo <b>int64</b> o <b>float64</b>. Una excelente manera de visualizar estas variables es mediante el uso de diagramas de dispersión con líneas ajustadas.</p>

Para poder visualizar los gráficos se van a importar las librerías Matplotlib y Seaborn. Se les pone un alias <strong>plt</strong> y <strong>sns</strong> para que sea más fácil su uso:

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

Se pueden utilizar varios tipos de gráficos para visualizar la información. A continuación se utiliza el gráfico <code>seaborn.pairplot</code> que permite visualizar todas las variables numéricas y la combinación entre ellas. Pudiendo identificar facilmente si existen relaciones de dependencia entre algunas variables.

```python
sns.pairplot(df)
```

Si se desean visualizar dos variables en un solo diagrama para mayor entendimiento y mejor visualización se puede utilizar un gráfico de <code>seaborn.regplot</code>. Se muestran los atributos de horas de trabajo por semana (<b>hours-per-week</b>) y ganancias de capital (<b>capital-gain</b>).

```python
#el parámetro style="whitegrid" permite poner fondo blanco con un grid
sns.set(style="whitegrid")
sns.regplot(x="hours-per-week", y="capital-gain", data=df)
```

Estos tipos de gráficos permiten identificar relaciones lineales entre los datos. El conjunto de datos de Adult no presenta relaciones lineales entre sus variables numéricas. A continuación se muestran ejemplos de otros gráficos donde si se aprecian relaciones de dependencia entre las variables.

<b>Variables categóricas</b>

<p>Estas son variables que describen una ‘característica’ de una unidad de datos y se seleccionan de un pequeño grupo de categorías. Las variables categóricas pueden tener el tipo <b>objeto</b> o <b>int64</b>. Una buena forma de visualizar variables categóricas es mediante el uso de diagramas de barra.</p>

Para realizar un gráfico sencillo que solo muestre una variable, se puede utilizar <code>sns.countplot</code>. De este gráfico se puede interprestar que hay 2 categorías únicas por género. La frecuencia de los hombres es más alta que la de las mujeres

```python
sns.countplot(x="sex", data=df,)
```

Se grafica ahora la variable <b>marital-status</b> utilizando otro tipo de gráfico en este caso un <code>sns.catplot</code>. De este gráfico se puede interpretar que la categoría <b>Married-civ-spouse</b> tiene el número máximo de muestras y la categoría <b>Married-AF-spouse</b> tiene un número mínimo de observaciones.

```python
sns.set(style="whitegrid")
sns.catplot(x="marital-status", kind="count", data=df, height = 8, aspect = 3)
```

Una vez que se hayan visualizado las variables de manera independiente, se pueden utilizar dos variables en el mismo gráfico para ver su relación. En este caso se visualizan las variables <b>workclass</b> y <b>makes</b> analizando la relación entre las dos categorías. 
Los datos parecen consistir principalmente en empleados privados y en todas las clases de trabajo, el número de personas que ganan menos de 50.000 son más que las que ganan 50.000.

```python
fig = plt.figure(figsize=(15,8))
sns.countplot(x='workclass',hue ='makes',data=df).set_title("workclass count VS makes")
```

Con el gráfico de <code>sns.catplot</code> se puede visualizar más de un gráfico a la vez. En esta visualización se combinan tres atributos <b>race</b>, <b>sex</b> y <b>make</b>. Se puede apreciar que las personas de género masculino y de raza blanca son los que más ganan ingresos por encima de 50mil.

```python
plt.figure(figsize=(25,10))
sns.catplot(y="race", hue="makes", kind="count",col="sex", data=df);
```
